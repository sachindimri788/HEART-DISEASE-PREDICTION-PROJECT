```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x=heart_data.drop(columns='target',axis=1)
y=heart_data['target']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,stratify=y,random_state=2)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train) 
```

```python
x_train_prediction=model.predict(x_train)
trianing_data_aaccuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy on training:", trianing_data_aaccuracy)
```

```python
x_test_prediction=model.predict(x_test)
test_data_aaccuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy on test:", test_data_aaccuracy)
```
